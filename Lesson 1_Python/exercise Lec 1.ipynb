### Question 04: Write a program that prints the first N (N>=2) numbers of the Fibonacci sequence. The Fibonacci sequence starts with 1 and 1, and each subsequent number is the sum of the two preceding ones like the following sequence:

```python
    1, 1, 2, 3, 5, 8, 13, ...
```

- If the input N is 5, you should print the result: 1, 2, 2, 3, 5.
- Your program should be able to output the result of any positive integer N.

### Write your answer in the following code frame:

````python
# Python program to calculate the area of a circle --by MSY

# Promote the user to input the number N
N = int(input("Please enter a random positive integer"))

# Fibonacci Sequence
if N == 1:
    print("1")
else:
    # Start the sequence with the first two numbers
    fibonacci_sequence =[1,1]
    # Generate the sequence up to the Nth number
    while len(fibonacci_sequence) < N:
        next_value =fibonacci_sequence[-1]+fibonacci_sequence[-2]
        fibonacci_sequence.append(next_value)
    # Print the sequence
    print(",".join(map(str, fibonacci_sequence)))
````

### Question 05: Define a function that calculates the factorial of a given non-negative integer. The factorial is defined as: 
$$n! = n (n-1) (n-2) \cdots 1$$

- Hint: You can choose to use recursive methods to define (not neccessary)
  
### Write your answer in the following code frame:


````python
# Python program to define a function --by MSY
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
````

### Question 10: Create a pandas DataFrame from the given dictionary (using the code) to store information about students, including their names, ages, and scores. Then, Calculate the average score of all students and Add a new column in the dataframe to store the grade (A, B, C, etc.) for each student based on their score.

```python
    import pandas as pd
    data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [20, 21, 19, 22],
    'Score': [83, 95, 73, 92]
    }
```

- The grading rules are: A (grade>=90), B (90>grade>=80),C (>80grade>=70), D(grade<70)

### Write your answer in the following code frame:

````python
# Python program to create and calculate a dictionary --by MSY
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [20, 21, 19, 22],
    'Score': [83, 95, 73, 92]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate average score
average_score = df['Score'].mean()

# Add grade column
df['Grade'] = pd.cut(df['Score'], bins=[0, 70, 80, 90, 100], labels=['D', 'C', 'B', 'A'])

print("Student Information:")
print(df)
print()

print("Average Score:", average_score)
````

### Question 11: Define a function that takes a sentence as input and returns the sentence with the first letter of each word capitalized. 

- Sample sentence: Today is a sunny day.

- Expected output: Today Is A Sunny Day.

### Write your answer in the following code frame:


````python
# Python program to define a function --by MSY
def capitalize_sentence(sentence):
    words = sentence.split()
    capitalized_words = [word.capitalize() for word in words]
    capitalized_sentence = ' '.join(capitalized_words)
    return capitalized_sentence

# Example usage
sentence = "Today is a sunny day."
capitalized_sentence = capitalize_sentence(sentence)
print(capitalized_sentence)
````

### Question 12 (optional): Define a class Rectangle with attributes width and height. Implement methods get_area() to calculate the area of the rectangle and get_perimeter() to calculate the perimeter. You Should use `input` function to receive the width and height values from users 


### Write your answer in the following code frame:

````python
# Python program to define a class --by MSY
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def get_area(self):
        return self.width * self.height

    def get_perimeter(self):
        return 2 * (self.width + self.height)

# Example usage
width = float(input("Enter the width of the rectangle: "))
height = float(input("Enter the height of the rectangle: "))

rectangle = Rectangle(width, height)

area = rectangle.get_area()
perimeter = rectangle.get_perimeter()

print("Area:", area)
print("Perimeter:", perimeter)
````
